```python
fig = make_subplots(rows=3, cols=1,
                    shared_xaxes=False,
                    vertical_spacing=.1,
                    subplot_titles=("Afluencia total preliminar","Afluencia boleto", "Afluencia tarjeta"),
                   )
                
fig.add_trace(
    go.Histogram(x=df["afluencia_tarjeta"],nbinsx =40, name="Afluencia tarjeta",),
              row=3, col=1)

fig.add_trace(
    go.Histogram(x=df["afluencia_boleto"],nbinsx =40, name="Afluencia boleto"),
              row=2, col=1)

fig.add_trace(
    go.Histogram(x=df["afluencia_total_preliminar"],nbinsx =40, name="Afluencia total preliminar" ),
              row=1, col=1)

fig.update_layout(height=900, width=1000,
                  title_text="Distribución de las columnas de afluencia")


fig
```
